# Step 2_Movies Promotion Behaviours Clustering
**Note**: This part of codes include 2 parts:
* Part I: Movies Promotion Behaviours Clustering in each cluster  
This part shows the clustering process and result of promotion behaviours clustering in each cluster we got in step 1.
* Part II: Explore the characteritics of each cluster  
This part shows the visualization of the each clusters' promotion behaviours for each sub-cluster. Through the visualization, we can conclude the promotion differences in each cluster.

**Data Required**:
movies_with_clusters.csv, f0_promotion.csv, f0_perform.csv, f1_promotion.csv, f1_perform.csv, f2_promotion.csv, f2_perform.csv

# Part 0: Import packages
Whenever you run any part of the codes, run the import part first.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.cluster import KMeans
from sklearn.preprocessing import MinMaxScaler
```

# Part I: Movies Promotion Behaviours Clustering in each cluster

```python
# read the data with feature clusters
movies_with_clusters = pd.read_csv('movies_with_clusters.csv')

# get the cluster with 'feature clusters == 0'
f0 = movies_with_clusters[movies_with_clusters.feature_clusters == 0]
# select the varibales about promotion behaviours
f0_promotion = f0.iloc[:,5:54]
# select the varibales about movies performances
f0_perform = f0.drop(list(f0_promotion.columns)+['release_time','duration', 'out_week','feature_clusters'], axis=1)

# get the cluster with 'feature clusters == 1'
f1 = movies_with_clusters[movies_with_clusters.feature_clusters == 1]
f1_promotion = f1.iloc[:,5:54]
f1_perform = f1.drop(list(f1.columns[5:54])+['release_time','duration', 'out_week','feature_clusters'], axis=1)

# get the cluster with 'feature clusters == 2'
f2 = movies_with_clusters[movies_with_clusters.feature_clusters == 2]
f2_promotion = f2.iloc[:,5:54]
f2_perform = f2.drop(list(f2.columns[5:54])+['release_time','duration', 'out_week','feature_clusters'], axis=1)
```

## 1.1 For movies in feature cluster 0

### (1) Find the optimal cluster number `k`

```python
wcss=[]
for i in range(1,7):
    kmeans = KMeans(i)
    kmeans.fit(f0_promotion)
    wcss_iter = kmeans.inertia_
    wcss.append(wcss_iter)

number_clusters = range(1,7)
plt.plot(number_clusters, wcss)
plt.title('The Elbow title')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
```

### (2) Clustering

```python
# data Normalization and clustering via KMeans
ms = MinMaxScaler()
km = KMeans(n_clusters = 2, random_state = 0)
f0_clusters = km.fit_predict(ms.fit_transform(f0_promotion))

# Add the cluster column
f0['promotion_clusters'] = f0_clusters
f0_promotion['promotion_clusters'] = f0_clusters
f0_perform['promotion_clusters'] = f0_clusters
```

```python
f0_promotion.to_csv("f0_promotion.csv", index = False)
f0_perform.to_csv("f0_perform.csv", index = False)
```
